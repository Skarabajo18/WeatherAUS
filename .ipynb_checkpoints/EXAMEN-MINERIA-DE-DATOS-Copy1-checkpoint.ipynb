Acá se rellenan los valores vacíos con algún valor, puede ser el promedio del valor de la columna. 
Puede ser el valor más repetido en el caso de una variable categórica.
Una extensión de la imputación. Se agrega una columna indicando que el valor ha sido reemplazado, de esta forma, se mantiene “identificados” a los valores nulos

# Transformacion de datos

```python
# Cambio de RainToday and RainTommorrow; No = 0, Yes: 1 // transforma la informacion - No = 0 y 'Yes' = 1
data_frame['RainTodayNumeric'] = data_frame['RainToday'].map( {'No':0, 'Yes':1} )
data_frame['RainTomorrowNumeric'] = data_frame['RainTomorrow'].map( {'No':0, 'Yes':1} )
```

```python
# Variable que muestra en un array las columnas que tienen datos nulos
# columns_NaN = data_frame.columns[data_frame.isnull().any()]
# columns_NaN
columns_NaN = []
columnas = data_frame.columns
for col in columnas:
    if(data_frame[col].dtypes != 'object'):
        if data_frame[col].isnull().any() != False:
            columns_NaN.append(col)
columns_NaN
```

```python
for column in columns_NaN:
    data_frame[column].fillna(data_frame[column].mean(), inplace=True)
```

# Valores nulos despues de la transformacion de datos

```python
#
data_frame['WindGustDirNotNull'] = data_frame['WindGustDir']
data_frame['WindDir9amNotNull'] = data_frame['WindDir9am']
data_frame['WindDir3pmNotNull'] = data_frame['WindDir3pm']

data_frame['WindGustDirNotNull'].fillna('W', inplace=True)
data_frame['WindDir9amNotNull'].fillna('N', inplace=True)
data_frame['WindDir3pmNotNull'].fillna('SE', inplace=True)
```

# Comparando la temperatura mínima y máxima de Katherine  2013 al 2017 

```python
nombre = ['Katherine2013', 'Katherine2014', 'Katherine2015', 'Katherine2016', 'Katherine2017']
anos = ['2013','2014','2015','2016','2017']


primerFiltro = data_frame[
    (data_frame['Location'].str.contains('Katherine') == True)
]
lineas = []
for i, j in zip(nombre, anos):
    temp = primerFiltro[
        (primerFiltro['Date'].str.contains(j) == True)
    ]
    lineas.append(temp)
    
    
```

```python
# primerFiltro
len(lineas)
```

```python
for i in lineas:
    fig, ax = plt.subplots()
    ax.plot(i['Date'], i['MinTemp'], color = 'tab:green')
    ax.plot(i['Date'], i['MaxTemp'], color = 'tab:red')
    plt.title("Distribución de Temperatura Max y Min")
plt.show()
```

# Comparando Temperatura Min y Max de Alice Springs

```python
nombre = ['AliceSprings2013']
anos = ['2013']

SegundoFiltro = data_frame[
    (data_frame['Location'].str.contains('AliceSprings') == True)
]
lineasS = []
for i, j in zip(nombre, anos):
    temp = SegundoFiltro[
        (SegundoFiltro['Date'].str.contains(j) == True)
    ]
    lineasS.append(temp)
for i in lineasS:
    fig, ax = plt.subplots()
    ax.plot(i['Date'], i['MinTemp'], color = 'tab:Blue')
    ax.plot(i['Date'], i['MaxTemp'], color = 'tab:Red')
    plt.title("2013")
plt.show()
```

```python
nombre = ['AliceSprings2014']
anos = ['2014']

SegundoFiltro = data_frame[
    (data_frame['Location'].str.contains('AliceSprings') == True)
]
lineasS = []
for i, j in zip(nombre, anos):
    temp = SegundoFiltro[
        (SegundoFiltro['Date'].str.contains(j) == True)
    ]
    lineasS.append(temp)
for i in lineasS:
    fig, ax = plt.subplots()
    ax.plot(i['Date'], i['MinTemp'], color = 'tab:Blue')
    ax.plot(i['Date'], i['MaxTemp'], color = 'tab:Red')
    plt.title("2014")
plt.show()
```

```python
#Demostracion de la relacion entre la Tem. Max,  Nubocidad y Humedad a las 9 AM
#En donde podemos apreciar que a mayor temperatura, mayor es la evaporación del agua en el ambiente y menor es la nubocidad relativa

groupad1=data_frame.groupby("Location").agg({
    "Temp9am":'mean',
    "Cloud9am":'mean',
    "Evaporation":'mean'
   })
groupad1.head().round()#round() para disminuir los decimales 
groupad1.head(10).plot(figsize=(20,5), kind = 'bar');
plt.title("Relacion entre la Tem. Max,  Nubocidad y Evaporación a las 9 AM")
```

```python
#Agrupamiento  Mínima Temperatura  y Maxima Temperaturaen relacion a la localidad

groupad1=data_frame.groupby("Location").agg({
    "Cloud9am":'max',
    "Cloud3pm":'max'
   })
groupad1.round()#round() para disminuir los decimales 
groupad1.plot(figsize=(20,5), kind = 'bar');
plt.title("Temperaturas Máximas a las 9AM y 3PM")
```
